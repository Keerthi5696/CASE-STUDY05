# 5. Check and handle outliers in at least 3 columns in the dataset

Here , handling outliers in colums "Age" ,"Fare" & "SibSp".

>> 1. Age

```python
plt.boxplot(df['Age'])   
plt.title('Age Outlier Checking')
plt.show()
```

```python
df["Age"]
```

```python
Q1_Age = df['Age'].quantile(.25)
Q1_Age
```

```python
Q2_Age = df['Age'].quantile(.50)
Q2_Age
```

```python
Q3_Age = df['Age'].quantile(.75)
Q3_Age
```

Finding the Interquartile Range

```python
IQR_Age = Q3_Age-Q1_Age
IQR_Age
```

Finding lower Limit

```python
low_limit_Age = Q1_Age-1.5*IQR_Age
low_limit_Age
```

Finding Upper Limit

```python
up_limit_Age = Q3_Age+1.5*IQR_Age
up_limit_Age
```

Finding Outlier Data

```python
Age_outlier= df[(df['Age'] < low_limit_Age ) | (df['Age'] > up_limit_Age )]
Age_outlier
```

Replacing Outlier Value with Mean Value

```python
outlier_age = (df['Age']<low_limit_Age)
outlier_age = df.loc[outlier_age].index
df.loc[list(outlier_age ),'Age']=df['Age'].mean()
df
```
